```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('world_population.csv')
df.head()
```

```python
plt.figure(figsize=(15,6))

plt.subplot(1,2,1)
plt.hist(df['2022 Population'], bins=10, color='teal', edgecolor='black')
plt.title('HISTOGRAM OF COUNTRY POPULATION (2022)')
plt.xlabel('POPULATION COUNT')
plt.ylabel('NUMBER OF COUNTRIES')
plt.tight_layout()
plt.grid()

df_top10 = df.sort_values('2022 Population', ascending=False).head(10) ## Filtering top 10 countries

plt.subplot(1,2,2)
plt.bar(df_top10['Country/Territory'], df_top10['2022 Population'], color='salmon', edgecolor='black')
plt.title('TOP 10 MOST POPULOUS COUNTRIES IN 2022')
plt.xlabel('COUNTRY')
plt.ylabel('POPULATION COUNT')
plt.xticks(rotation=45)
plt.grid()
plt.tight_layout()

plt.show()
```

```python
year = ['2022 Population', '2020 Population', '2015 Population', '2010 Population', '2000 Population', '1990 Population', '1980 Population', '1970 Population']

plt.figure(figsize=(20,8))
plt.title('WORLD POPULATION FROM YEAR 1970-2022')

for idx,i in enumerate(year):
  plt.subplot(2,4,idx+1)
  df_top10 = df.sort_values(i, ascending=False).head(5)
  plt.bar(df_top10['Country/Territory'], df_top10[i], edgecolor='black', color='Purple')
  plt.ylabel('Population Count')
  plt.xlabel(i)
  plt.xticks(rotation=45)
  plt.grid()
  plt.tight_layout()

plt.show()
```

# CONCLUSION (According to old Dataset i.e from 1970-2022):-

- The histogram shows that most countries in the world have relatively small populations—almost all are clustered near the lower end of the scale. Only a very few countries (such as China, India, and the United States) have extremely large populations, which appear as isolated bars on the far right of the chart.

- China and India consistently rank as the two most populous countries in every decade analyzed, clearly dominating the global population landscape.

- Pakistan, Russia, Indonesia, and Brazil show fluctuation in the top 5 rankings of the most populous countries over the decades.
